```python
SalNorm_predict = []
SalNorm_real = []

#####
start_index = 244_750  # start with high number for testing, because lot of processing power&time needed
for n in range(start_index, len(df_val['SalaryNormalized'])):
    row = df_val.iloc[n]
    
    # filter by column value
    filt = df_train[(df_train['LocationNormalized'] == row['LocationNormalized'])&
                    (df_train['Category'] == row['Category'])] 
    
    # add predicted salaries
    mean_SalNorm = filt['SalaryNormalized'].mean()
    SalNorm_predict.append(mean_SalNorm)
    
    # add real salaries
    realSalNorm = row[['SalaryNormalized'][0]]
    SalNorm_real.append(realSalNorm)

    
# print predicted and real SalaryNormalized
for i in range(0, len(SalNorm_real)):
    print('predict:', SalNorm_predict[i], '\nreal:', SalNorm_real[i])

```

```python
difference = []
for n in range(0, len(SalNorm_predict)):
    dif = (SalNorm_predict[n] - SalNorm_real[n])
    difference.append(dif)
    
for n in difference:
    print(n)
```
