```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('train_rev1.csv')
```

```python
#transformation target feature
df[["SalaryNormalized"]] = df[["SalaryNormalized"]].apply(pd.to_numeric)
df['SalaryNormalized'] = df['SalaryNormalized'].apply(lambda x: round(x*0.001))
df
```

```python
## metropol feature
df['metropol']= df['LocationNormalized'].apply(lambda x: 1 if x=="London" else '0')

## big city feature

big_city= ["Birmingham","Glasgow","Leeds","Manchester"] # above 1 mil 
df['big_city']= df['LocationNormalized'].isin(big_city).astype(int) 

## medium city feature    

medium_cities = ["Liverpool","Sheffield","Newcastle","Southampton","Portsmouth","Belfast","Brighton","Worthing"
                ,"Littlehampton","Leicester","Edinburgh","Poole","Bournemouth"] # above 500.000 and smaler 1 mil 


df['big_city']= df['LocationNormalized'].isin(medium_cities).astype(int)


#small_cities  = []
# if not than small city
```

```python
## feature contract type and contract time 

## with filled empty cells (majority: full_time, permanent)
df.fillna({'ContractType':'full_time', 'ContractTime': "permanent"}, inplace=True)

## delete instances with empty slots
#df_work.dropna(subset=['ContractType', 'ContractTime'])
```
